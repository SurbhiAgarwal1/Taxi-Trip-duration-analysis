# TaxiIQ Dataset Exploration

This notebook provides code to load and explore the various dataset files used in the TaxiIQ project, such as trip records, zone lookups, and calculated metrics.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options to view more columns if needed
pd.set_option('display.max_columns', None)

# Define the path to the data directory
DATA_DIR = 'data'
```

## 1. Taxi Zone Lookup
This file maps Location IDs to specific Boroughs and Zones in NYC.

```python
zone_lookup_path = os.path.join(DATA_DIR, 'taxi_zone_lookup.csv')
if os.path.exists(zone_lookup_path):
    zone_lookup_df = pd.read_csv(zone_lookup_path)
    display(zone_lookup_df.head())
else:
    print(f"File not found: {zone_lookup_path}")
```

## 2. Cleaned Taxi Data
This dataset contains the processed and cleaned taxi trips.

```python
clean_taxi_path = os.path.join(DATA_DIR, 'taxi_clean.parquet')
if os.path.exists(clean_taxi_path):
    taxi_clean_df = pd.read_parquet(clean_taxi_path)
    display(taxi_clean_df.head())
    print(f"Total shape: {taxi_clean_df.shape}")
else:
    print(f"File not found: {clean_taxi_path}")
```

## 3. Raw Trip Data (Example: Yellow 2025-01)
Loading one of the raw monthly parquet files to inspect its structure.

```python
yellow_tripdata_path = os.path.join(DATA_DIR, 'yellow_tripdata_2025-01.parquet')
if os.path.exists(yellow_tripdata_path):
    yellow_trip_df = pd.read_parquet(yellow_tripdata_path)
    display(yellow_trip_df.head())
else:
    print(f"File not found: {yellow_tripdata_path}")
```

## 4. Pre-calculated Metrics (Corridor Metrics)
Loading aggregated metrics calculated across various zone corridors.

```python
corridor_metrics_path = os.path.join(DATA_DIR, 'corridor_metrics.parquet')
if os.path.exists(corridor_metrics_path):
    corridor_metrics_df = pd.read_parquet(corridor_metrics_path)
    display(corridor_metrics_df.head())
else:
    print(f"File not found: {corridor_metrics_path}")
```

## 5. Zone Summaries
Loading zone-level summary metrics.

```python
zone_summary_path = os.path.join(DATA_DIR, 'zone_summary.parquet')
if os.path.exists(zone_summary_path):
    zone_summary_df = pd.read_parquet(zone_summary_path)
    display(zone_summary_df.head())
else:
    print(f"File not found: {zone_summary_path}")
```
